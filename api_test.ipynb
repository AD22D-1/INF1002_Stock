## Import Libraries

```python
import requests
import pandas as pd
import time
from datetime import datetime
```

```python
# List of Selected Tickers we want to analyse 
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA']
```

```python
# store stockdata here 
stock_data = []
```

# API URL and Key 

```python
# The URL for monthly data
# Note: The function is TIME_SERIES_MONTHLY, and there is no "adjusted" data
api_key = "9JPRF2OKFQM0FT0U"
base_url_template = 'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol={symbol}&apikey={api_key}'
```

Requesting for multiple Stock Data

```python

for tick in tickers:
    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol={tick}&apikey={api_key}'
    try:
        r = requests.get(url)
        r.raise_for_status()  
        data = r.json()

        if "Error Message" in data:
            print(f"Error for {tick}: {data['Error Message']}")

        else:
            print(f"--- Data for {tick} ---")
            print(data)
    except requests.exceptions.RequestException as e:
        print(f"An error occurred while fetching data for {tick}: {e}")
```

## Storing data to Stock_Data
Transforming open, close, high, low from str to floats
Date to date object 
volume from str to int

```python
for tick in tickers:
    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol={tick}&apikey={api_key}'
    try:
        r = requests.get(url)
        r.raise_for_status()  
        data = r.json()

        if "Error Message" in data:
            print(f"Error for {tick}: {data['Error Message']}")

        else:
            print(f"--- Data for {tick} ---")
            print(data)
    except requests.exceptions.RequestException as e:
        print(f"An error occurred while fetching data for {tick}: {e}")
        
time_series_key = 'Monthly Time Series' # Depends on what key is returned by the specified API function 
if time_series_key in data:
    for date_str, values in data[time_series_key].items():
        row = {
            'Symbol': tick,
            'Date': date_str,  
            'Open': float(values['1. open']),
            'High': float(values['2. high']),
            'Low': float(values['3. low']),
            'Close': float(values['4. close']),
            'Volume': int(values['5. volume'])
        }
        stock_data.append(row)
    print(f"--- Successfully collected and formatted data for {tick} ---")
            
elif "Error Message" in data:
    print(f"API Error for {tick}: {data['Error Message']}")
else:
    print(f"Unexpected data format for {tick}.")
time.sleep(20)
```

Convert to Data Frame 

```python
stock_df = pd.DataFrame(stock_data)
```

```python
# convert date from str to date time
stock_df['Date'] = pd.to_datetime(stock_df['Date'])
```
